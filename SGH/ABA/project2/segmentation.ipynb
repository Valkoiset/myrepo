## 3. Clustering

#### 3.1 Inertia graph:

```python
sse={}
tx_recency = rfm[['Recency']]
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, max_iter=1000).fit(tx_recency)
    clusters = kmeans.labels_
    sse[k] = kmeans.inertia_ 
plt.figure()
plt.plot(list(sse.keys()), list(sse.values()))
plt.xlabel("Number of clusters")
plt.savefig(r"4_Number_of_clusters.png")
plt.show()
```

```python
# assigning clusters
n_clusters=4
```

#### 3.2 Helper functions

```python
#order cluster method
def order_cluster(cluster_field_name, target_field_name,df,ascending):
    new_cluster_field_name = 'new_' + cluster_field_name
    df_new = df.groupby(cluster_field_name)[target_field_name].mean().reset_index()
    df_new = df_new.sort_values(by=target_field_name,ascending=ascending).reset_index(drop=True)
    df_new['index'] = df_new.index
    df_final = pd.merge(df,df_new[[cluster_field_name,'index']], on=cluster_field_name)
    df_final = df_final.drop([cluster_field_name],axis=1)
    df_final = df_final.rename(columns={"index":cluster_field_name})
    return df_final
```

```python
def clustering(in_model, df_orig, df_in):    
    #dataframe
    df_out=df_orig.copy(deep=True)    
    
    # recency
    model = in_model
    model.fit(df_in[['Recency']])
    df_out['RecencyCluster'] = model.predict(df_in[['Recency']])
    df_out = order_cluster('RecencyCluster', 'Recency',df_out,False)
    
    # frequency
    model = in_model
    model.fit(df_in[['Frequency']])
    df_out['FrequencyCluster'] = model.predict(df_in[['Frequency']])
    df_out = order_cluster('FrequencyCluster', 'Frequency',df_out,True)
    
    # monetary
    model = in_model
    model.fit(df_in[['Monetary']])
    df_out['MonetaryCluster'] = model.predict(df_in[['Monetary']])
    df_out = order_cluster('MonetaryCluster', 'Monetary',df_out,True)
    
    #overall scoring
    df_out['OverallScore'] = df_out['RecencyCluster'] + df_out['FrequencyCluster'] + df_out['MonetaryCluster']
    df_out['Segment'] = 'Low-Value'
    df_out.loc[df_out['OverallScore']>2,'Segment'] = 'Mid-Value' 
    df_out.loc[df_out['OverallScore']>4,'Segment'] = 'High-Value'     
    return df_out
```

#### 3.3 Quartile clustering

```python
# dataset
quartile = pd.DataFrame(rfm).copy(deep=True)

#clustering
quartile['RecencyCluster'] = pd.qcut(normalized_rfm['Recency'], 4, ['3','2','1','0']).astype(int)
quartile['FrequencyCluster'] = pd.qcut(normalized_rfm['Frequency'], 4, ['0','1','2','3']).astype(int)
quartile['MonetaryCluster'] = pd.qcut(normalized_rfm['Monetary'], 4, ['0','1','2','3']).astype(int)

#overall scoring
quartile['OverallScore'] = quartile['RecencyCluster'].astype(int) + quartile['FrequencyCluster'].astype(int) + quartile['MonetaryCluster'].astype(int)
quartile['Segment'] = 'Low-Value'
quartile.loc[quartile['OverallScore']>2,'Segment'] = 'Mid-Value' 
quartile.loc[quartile['OverallScore']>4,'Segment'] = 'High-Value' 

quartile
```

#### 3.4 K-means clustrering

```python
kmodel = KMeans(n_clusters=n_clusters)
kmeans = clustering(in_model=kmodel, df_orig=rfm, df_in=normalized_rfm)
kmeans
```

#### 3.5 BIRCH clustering

```python
bmodel = Birch(threshold=0.01, n_clusters=n_clusters)
birch = clustering(in_model=bmodel, df_orig=rfm, df_in=normalized_rfm)
birch
```

#### 3.6 Gaussian Mixture Model

```python
gmodel = GaussianMixture(n_components=n_clusters)
gaussian = clustering(in_model=gmodel, df_orig=rfm, df_in=normalized_rfm)
gaussian
```

## 5. Modelling future data

#### 5.1 Calculating revenue for future data

```python
#calculate revenue and create a new dataframe for it
uk_data_6m['Revenue'] = uk_data_6m['UnitPrice'] * uk_data_6m['Quantity']
clv = uk_data_6m.groupby('CustomerID')['Revenue'].sum().reset_index()
clv.columns = ['CustomerID','Revenue']
```

#### 5.2 Helper functions

```python
def grouping(df,model):
    # merge datasets together
    df_merge = pd.merge(df, clv, on='CustomerID', how='left')
    df_merge = df_merge.fillna(0)
    #remove outliers
    df_merge = df_merge[df_merge['Revenue']<df_merge['Revenue'].quantile(0.99)]

    #creating 3 clusters
    model.fit(df_merge[['Revenue']])
    df_merge['LTVCluster'] = model.predict(df_merge[['Revenue']]).astype(int)

    #order cluster number based on LTV
    df_merge = order_cluster('LTVCluster', 'Revenue',df_merge,True)

    #convert categorical columns to numerical
    df_merge = pd.get_dummies(df_merge)
    return df_merge
```

```python
def xgb_model(df):
    #create X and y, X will be feature set and y is the label - LTV
    X = df[['RecencyCluster','FrequencyCluster','MonetaryCluster']]    #.drop(['LTVCluster','Revenue'],axis=1)
    y = df['LTVCluster']

    #split training and test sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.05, random_state=56)

    #XGBoost Multiclassification Model
    ltv_xgb_model = xgb.XGBClassifier(max_depth=5, learning_rate=0.1,objective= 'multi:softprob',n_jobs=-1).fit(X_train, y_train)

    print('Accuracy of XGB classifier on training set: {:.2f}'
           .format(ltv_xgb_model.score(X_train, y_train)))
    print('Accuracy of XGB classifier on test set: {:.2f}'
           .format(ltv_xgb_model.score(X_test[X_train.columns], y_test)))
```

#### 5.3 Quantile estimation

```python
# merge datasets together
q_df = pd.merge(quartile, clv, on='CustomerID', how='left')
q_df = q_df.fillna(0)
#remove outliers
q_df = q_df[q_df['Revenue']<q_df['Revenue'].quantile(0.99)]

#creating 3 clusters
q_df['LTVCluster'] = pd.qcut(q_df['Revenue'], 3, ['2','1','0']).astype(int)

#convert categorical columns to numerical
q_df = pd.get_dummies(q_df)

xgb_model(q_df)
```

```python
kmodel = KMeans(n_clusters=3)
k_df=grouping(kmeans,kmodel)
xgb_model(k_df)
```

#### 5.5 Birch estimation

```python
bmodel = Birch(threshold=0.01, n_clusters=3)
b_df=grouping(birch,bmodel)
xgb_model(b_df)
```

```python
gmodel = GaussianMixture(n_components=3)
g_df=grouping(gaussian,gmodel)
xgb_model(g_df)
```
