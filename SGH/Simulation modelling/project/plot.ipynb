```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def smooth(x):
    # last 100
    n = len(x)
    y = np.zeros(n)
    for i in range(n):
        start = max(0, i - 99)
        y[i] = float(x[start:(i + 1)].sum()) / (i - start + 1)
    return y
```

# Pipe_gap = 125

- pop_size = 30; iterations = 150

```python
# training time - 23 minutes
j = np.load('./results/pipe_gap=125/30_150/es_flappy_results.npz')

returns = j['train'] + 5

plt.plot(returns)
plt.plot(smooth(np.array(returns)))
plt.title("Train returns. Training time: 00:23:14\nPipe gap = 125, pop.size = 30, iterations = 150")
plt.xlabel("Iterations")
plt.ylabel("Avg Reward")
plt.savefig('125_30_150.png')
plt.show()
```
