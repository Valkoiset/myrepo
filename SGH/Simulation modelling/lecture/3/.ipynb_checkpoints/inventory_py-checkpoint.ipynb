```python
import numpy as np
import numpy.random as rd

def simulateOneRun(m, s, S,
    h = 0.1, #storage cost
    c = 2.0, #unit profit
    p = 0.5, #probability of delivery
    K = 40, # fixed order cost
    k = 1.0, # variable order cost
    average_demand = 20):
                   
    Xj = S #Stock in the morning
    Yj = None #  and in the evening.
    profit = 0.0 # Cumulated profit.
    lost_orders = 0
    inventory_levs_history = []
    for j in range(m):
        Dj = rd.poisson(average_demand)
        Sj = min(Dj,Xj)
        lost_orders += max(0,Dj-Xj)
        Yj = Xj - Sj
        profit += c*Sj
        profit -= h*Yj
        if Yj <= s and rd.rand() < p:
            profit -= K
            profit -= k*(S-Yj)
            Yj = S
        Xj = Yj
        inventory_levs_history.append(Yj)
    return (profit, lost_orders, inventory_levs_history)
```

```python
%matplotlib inline
```

```python
rd.seed(0)
stany = simulateOneRun(100,10,400)[2]
import matplotlib.pyplot as plt
plt.plot(range(100),stany)
plt.show()
```

```python
def simulateManyRuns(N,m,s,S):
    profits = np.zeros(N)
    lost_os = np.zeros(N)
    for i in range(N):
        res = simulateOneRun(m,s,S)
        profits[i] = res[0]
        lost_os[i] = res[1]
    return {
              "profit":(np.mean(profits),
                        np.std(profits)),
              "lost_orders":(np.mean(lost_os),
                        np.std(lost_os))
            }
```

```python
rd.seed(0)
simulateManyRuns(100000,40,50,200)
```

```python
# let us check for various values of s and S
rd.seed(0)
results = {}
for s in range(10,350,40):
    for S in range(s+10,1050,40):
        results[(s,S)] = simulateManyRuns(2000,40,s,S)
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm

vals = [r for r in results.keys() if results[r]["profit"][0]>-500]
valsn = [r for r in results.keys() if results[r]["profit"][0]<=-500]

plt.scatter([v[1] for v in vals],
            [v[0] for v in vals], 
            c=[results[v]["profit"][0] for v in vals],
            cmap=cm.plasma_r )
plt.colorbar()
plt.scatter([v[1] for v in valsn],
            [v[0] for v in valsn],
            c="gray")
plt.xlabel("S - target wharehouse stock ")
plt.ylabel("s - reorder point")

plt.title("Profits")
plt.savefig(r"c:\temp\myplot.png")
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
vals = [r for r in results.keys() if results[r]["profit"][0]>-500]
valsn = [r for r in results.keys() if results[r]["profit"][0]<=-500]
plt.scatter([v[1] for v in vals],
            [v[0] for v in vals], 
            c=[results[v]["lost_orders"][0] for v in vals], cmap=cm.plasma_r )
plt.xlabel("S - target wharehouse stock ")
plt.ylabel("s - reorder point")
plt.colorbar()
plt.scatter([v[1] for v in valsn],
            [v[0] for v in valsn],
            c="gray")
plt.title("Lost orders")
plt.show()
```

```python
best_key = None
best_profit = 0
for key in results.keys():
    if results[key]["profit"][0] > best_profit:
       best_profit =  results[key]["profit"][0]
       best_key = key
       

print("Best decision: " ,best_key)
print("Best outcome:", results[best_key])
```
